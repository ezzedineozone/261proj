# An NLP Project for sentiment analysis
in this project we will be using the dataset from Kaggle to build a sentiment analysis model using NLP techniques. The dataset consists of text statements representing an individual’s thoughts or emotions, and their corresponding mental health status labels. The data is sourced from diverse platforms including social media posts, Reddit posts, Twitter posts, and more. Each statement is tagged with one of the following seven categories:

```python
import pandas as pd 
import numpy as np

df = pd.read_csv('Sentiment_analysis_dataset.csv')
# make df 30%
df = df.sample(frac=0.3, random_state=1)
```

## DATA PREPROCESSING

In this section, we will be doing data preprocessing steps to remove unwanted text that is irrelevant to our analysis. such as:
- Removing special characters
- Removing stop words
- Removing numbers
- Removing extra spaces
- Removing URLs
- Removing HTML tags
- Removing emojis
- Removing punctuation
- Removing non-ASCII characters



```python
import re
import string
from nltk.corpus import words
from nltk import download

# Download the English words corpus
download('words')
english_words = set(words.words())

def clean_text(text):
    if not isinstance(text, str):
        return ''
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    stop_words = set(['the', 'is', 'in', 'and', 'to', 'a'])
    text = ' '.join([word for word in text.split() if word not in stop_words])
    text = ''.join([i for i in text if not i.isdigit()])
    text = ' '.join(text.split())
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'<.*?>', '', text)
    text = text.encode('ascii', 'ignore').decode('ascii')
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = ''.join([i for i in text if ord(i) < 128])
    text = ' '.join([word for word in text.split() if word.lower() in english_words])
    
    return text

df['Statement'] = df['Statement'].astype(str).apply(clean_text)
```

```python

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk import download
download('punkt')
download('stopwords')
download('omw-1.4')
download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()


download('punkt_tab')




def preprocess_text(text):
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word.isalpha()]
    tokens = [word.lower() for word in tokens] 
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return tokens 
df['tokens'] = df['Statement'].apply(preprocess_text)
```

## Now that we have our tokens, we can start creating our features

```python
# first feature, BoW
# example from tokens column [life, doesnt, feel, worth, it, thats, kind]
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(analyzer=lambda x: x, max_features=5000)
X = vectorizer.fit_transform(df['tokens'])
X = X.toarray()
# convert to DataFrame
X = pd.DataFrame(X, columns=vectorizer.get_feature_names_out())
df['tokens'] = df['tokens'].apply(lambda x: x if isinstance(x, list) else [])
# add the target variable
Y = df['Status']

```

```python
# X is the feature matrix and Y is the target variable
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
# Train a Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
# Make predictions
y_pred = rf.predict(X_test)
# Evaluate the model
from sklearn.metrics import classification_report, accuracy_score
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
import pandas as pd
import numpy as np
import re
import nltk
from textblob import TextBlob
from nltk.corpus import opinion_lexicon, stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import Counter
nltk.download('opinion_lexicon')
positive_words = set(opinion_lexicon.positive())
negative_words = set(opinion_lexicon.negative())
stop_words = set(stopwords.words('english'))
```

```python
def get_sentiment_percentages(text):
    tokens = word_tokenize(text.lower())
    words = [word for word in tokens if word.isalpha()]
    total_words = len(words)
    if total_words == 0:
        return 0, 0
    pos_count = sum(1 for word in words if word in positive_words)
    neg_count = sum(1 for word in words if word in negative_words)
    return pos_count / total_words, neg_count / total_words

df[['pos_percent', 'neg_percent']] = df['Statement'].apply(
    lambda x: pd.Series(get_sentiment_percentages(x))
)
df.head()
```

```python

def get_negative_verb_percentage(text):
    tokens = word_tokenize(text.lower())
    words = [word for word in tokens if word.isalpha()]
    total_words = len(words)
    if total_words == 0:
        return 0
    neg_count = sum(1 for word in words if word in negative_words)
    return neg_count / total_words
df['neg_verb_percent'] = df['Statement'].apply(get_negative_verb_percentage)
df.head()
```

```python
mental_health_terms = set(['depressed', 'anxious', 'panic', 'suicidal', 'worthless', 'hopeless', 'tired', 'lonely', 'empty', 'overwhelmed', 'stressed'])

def mental_health_term_percent(text):
    tokens = word_tokenize(text.lower())
    words = [word for word in tokens if word.isalpha()]
    if not words:
        return 0
    count = sum(1 for word in words if word in mental_health_terms)
    return count / len(words)
df['mental_health_term_percent'] = df['Statement'].apply(mental_health_term_percent)
```

```python
def word_count(text):
    return len(word_tokenize(text))
df['word_count'] = df['Statement'].apply(word_count)
```

```python
# add subjectivity score column
def subjectivity_score(text):
    return TextBlob(text).sentiment.subjectivity
df['subjectivity_score'] = df['Statement'].apply(subjectivity_score)
```

```python
from nltk.tokenize import word_tokenize
from nltk import download
download('punkt')
first_person_pronouns = {"i", "me", "my", "mine", "we", "us", "our", "ours"}
second_person_pronouns = {"you", "your", "yours"}
third_person_pronouns = {"he", "him", "his", "she", "her", "hers", "it", "its", "they", "them", "their", "theirs"}

def calculate_pronoun_ratios(text):
    tokens = word_tokenize(text.lower())
    total_words = len(tokens)
    if total_words == 0:
        return 0, 0, 0

    first_person_count = sum(1 for word in tokens if word in first_person_pronouns)
    second_person_count = sum(1 for word in tokens if word in second_person_pronouns)
    third_person_count = sum(1 for word in tokens if word in third_person_pronouns)

    return (
        first_person_count / total_words,
        second_person_count / total_words,
        third_person_count / total_words,
    )
df[['first_person_ratio', 'second_person_ratio', 'third_person_ratio']] = df['Statement'].apply(
    lambda x: pd.Series(calculate_pronoun_ratios(x))
)
```

```python

from nltk.corpus import sentiwordnet as swn

def calculate_emotion_shift(text):
    tokens = word_tokenize(text.lower())
    words = [word for word in tokens if word.isalpha()]
    total_words = len(words)
    if total_words == 0:
        return 0
    emotion_counts = Counter()
    for word in words:
        synsets = list(swn.senti_synsets(word))
        if synsets:
            synset = synsets[0]
            emotion_counts[synset.synset.name().split('.')[0]] += 1
    return sum(emotion_counts.values()) / total_words
df['emotion_shift'] = df['Statement'].apply(calculate_emotion_shift)
```

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer

# Join tokens back into strings for CountVectorizer


# Vectorize
vectorizer = CountVectorizer(max_df=0.9, min_df=2, stop_words='english')
doc_term_matrix = vectorizer.fit_transform(df['Statement'])

# Fit LDA
lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda.fit(doc_term_matrix)

# Assign dominant topic
df['topic'] = lda.transform(doc_term_matrix).argmax(axis=1)
# print lda results

```

```python
print('LDA Topics:')
for index, topic in enumerate(lda.components_):
    print(f"Topic {index}:")
    print([vectorizer.get_feature_names_out()[i] for i in topic.argsort()[-10:]])

print(df['topic'].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode the target variable
label_encoder = LabelEncoder()
Y = label_encoder.fit_transform(df["Status"])  # Convert string labels to numeric

X = df.drop(columns=["Status", "Statement", "tokens"])

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
# use xgboost
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report

# Train the model
model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
model.fit(X_train, y_train)
# Make predictions
y_pred = model.predict(X_test)
# Evaluate the model
print(classification_report(label_encoder.inverse_transform(y_test), label_encoder.inverse_transform(y_pred)))
print("Accuracy:", accuracy_score(label_encoder.inverse_transform(y_test), label_encoder.inverse_transform(y_pred)))
```

```python
# print feature importances
import matplotlib.pyplot as plt
import seaborn as sns
importances = model.feature_importances_
feature_names = X.columns
indices = np.argsort(importances)[::-1]
plt.figure(figsize=(10, 6))
plt.title("Feature importances")
plt.bar(range(X.shape[1]), importances[indices], align="center")
plt.xticks(range(X.shape[1]), feature_names[indices], rotation=90)
plt.xlim([-1, X.shape[1]])
plt.show()




```

```python
# use neural networks to save the model, use grid search to find the best parameters
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
# Create a pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('mlp', MLPClassifier(max_iter=1000, random_state=42))
])
# Define the parameter grid
param_grid = {
    'mlp__hidden_layer_sizes': [(50,), (100,), (50, 50)],
    'mlp__activation': ['tanh', 'relu'],
    'mlp__alpha': [0.0001, 0.001],
    'mlp__learning_rate_init': [0.001, 0.01]
}
# Create a GridSearchCV object
grid_search = GridSearchCV(pipeline, param_grid, cv=3, n_jobs=-1, verbose=2)
# Fit the model
grid_search.fit(X_train, y_train)
# Get the best parameters
best_params = grid_search.best_params_


# use best_params to train the model
# Remove the 'mlp__' prefix from the keys in best_params
mlp_params = {key.replace('mlp__', ''): value for key, value in best_params.items()}
best_model = MLPClassifier(**mlp_params, max_iter=1000, random_state=42)
best_model.fit(X_train, y_train)
# Make predictions
y_pred = best_model.predict(X_test)
# Evaluate the model
print(classification_report(label_encoder.inverse_transform(y_test), label_encoder.inverse_transform(y_pred)))
print("Accuracy:", accuracy_score(label_encoder.inverse_transform(y_test), label_encoder.inverse_transform(y_pred)))

```
